Question 1: Create a new column in the provided dataset. Column name should be num_of_cast.
    The assignment was already explained in class. You have to append or add a column in the dataset which
    contains the total number of cast in that particular movie or Tv-Show.

```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("netflix_data.csv")
data.head()
```

```python
data.tail()
```

```python
 def cast_len(val):
    if val == "None":
        return 0
    else:
        return len(str(val).split(","))
```

```python
data["cast"].isnull().sum()
```

```python
data["cast"].replace(np.nan,"None",inplace=True)
```

```python
data.tail()
```

```python
data["cast_length"]=data["cast"].apply(cast_len)
```

```python
x=np.linspace(0,10,1000)
plt.plot(x, np.sin(x),color='b')
plt.plot(x, np.sin(x-1),color='g')
plt.plot(x, np.sin(x-2),color='#000')
plt.plot(x, np.sin(x-3),color='r')
plt.plot(x, np.sin(x-4),color='chartreuse')
plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x, np.sin(x),color='b', linestyle="solid")
plt.plot(x, np.sin(x-1),color='g',  linestyle="dashed")
plt.plot(x, np.sin(x-2),color='#000',  linestyle="dotted")
plt.plot(x, np.sin(x-3),color='r',  linestyle="dashdot")
plt.plot(x, np.sin(x-4),color='chartreuse', linestyle="-.")

plt.show()
```

```python
x=np.linspace(0,10,1000)
plt.plot(x, np.sin(x),color='b' , label="Sin(x)")
plt.plot(x, np.sin(x-1),color='g',label="Sin(x-1)")
plt.plot(x, np.sin(x-3),color='r', label="Sin(x-2)")

plt.title("Sign Graph")
plt.xlabel("x")
plt.ylabel("sin(x)")

plt.legend()
plt.show()
```
